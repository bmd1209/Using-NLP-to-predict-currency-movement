```python
data = pd.read_csv('ada_df.csv')
data
```

```python
data['log_diff'] = np.log(data['Close']) - np.log(data['Close'].shift(1))
data
```

```python
data['target'] = [1 if log_diff > 0 else 0 for log_diff in data['log_diff']]
data
data.to_csv(r'ADA_prices_log.csv')

```

```python
format = '%d-%m-%Y %H:%M:%S'
timezone = 'GMT'
data['timestamp'] = OpenBlender.dateToUnix(data['Date'], date_format = format, timezone = timezone)

data = data[['Date', 'timestamp', 'Close', 'target']]
data.to_csv(r'ADA_prices_target.csv')
```
